```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame({'x':np.random.rand(100),
                   'y':np.random.normal(0.0, 1.0, 100),
                   'cat':np.random.choice(['A','B','C'], 100)})
df.head(20)
```

```python
ax = sns.scatterplot(x='x', y='y', data=df)     # seaborn 0.9
```

```python
ax = sns.regplot(x='x', 
                 y='y', 
                 data=df, 
                 fit_reg=False, 
                 marker='o',
                 color='coral',
                 scatter_kws={'s': 15})

plt.ylim(0, 20)
plt.xlim(0, 1)
```

```python
ax = sns.lmplot(x='x', y='y', col='cat', data=df)
```
