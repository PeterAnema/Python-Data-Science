```python
from PIL import Image, ImageFilter
```

```python
try:
    original = Image.open("images/stinkbug.png")
#     original = Image.open("images/lena512.bmp")
except:
    print('Unable to load image')
```

```python
original.show()
```

```python
print("The size of the Image is: ")
print(original.format, original.size, original.mode)
```

```python
# Blur the image
blurred = original.filter(ImageFilter.BLUR)

# Display both images
blurred.show()

# save the new image
blurred.save("blurred.png")
```

```python
filtered = original.filter(ImageFilter.FIND_EDGES)
filtered.show()
```

```python
original.thumbnail((128, 128))
original.show()
```
