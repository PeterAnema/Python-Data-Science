```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Subplots

```python
x = np.linspace(0, 10, 100)

y1 = np.exp(-x)
y2 = np.exp(-(x**2))
y3 = np.exp(-(x**3))
y4 = np.exp(-(x**4))

fig,ax = plt.subplots(2,2,sharex=True,sharey=True)

ax[0][0].plot(x,y1)
ax[0][0].spines['bottom'].set_position('zero')
ax[0][0].spines['left'].set_position('zero')

ax[0][1].plot(x,y2)
ax[1][0].plot(x,y3)
ax[1][1].plot(x,y4)

# plt.subplot(221)
# plt.plot(x,y1)

# plt.subplot(222)
# plt.plot(x,y2)
# plt.plot(x,y1)

# plt.subplot(223)
# plt.plot(x,y3)

# plt.subplot(224)
# plt.plot(x,y4)
# plt.plot(x,y1)
```

### Line plot

```python
a = np.linspace(0, 10., 9)
b = np.exp(-a)
plt.plot(a, b)
plt.plot(a,a/10., linestyle='dotted', marker='D')    # or ':D'
plt.show()
```

### Histogram

```python
x = np.random.normal(size=2000, loc=5, scale=10)
plt.hist(x, bins=20)
plt.show()
```

### Bar plot

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

fig, (ax1, ax2) = plt.subplots(2,1)
data.plot.bar(ax=ax1)

data.plot.barh(ax=ax2)
```

### Scatter plot

```python
a = np.random.rand(100)
b = np.random.rand(100)
plt.scatter(a, b, alpha=0.5, color='red')
plt.scatter(b, a, alpha=0.5, color='green')
plt.show()
```

### 3D plot

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
plt.show()
```

### Contour plot

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)

Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

plt.contour(X, Y, Z)
```

###  Polar plots

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
