# Add a new column

```python
df['name'] = df['first_name'] + ' ' + df['last_name']
```

```python
df['name2'] = df[['first_name','last_name']].apply(lambda x:' '.join(x), axis=1)
```

```python
df['netnumber'] = df['phone1'].apply(lambda x: int(x[:3]))
```

```python
df['newby2'] = df['netnumber'] * 2
```

```python
df['newby3'] = df['netnumber'] + df['newby2']
```
